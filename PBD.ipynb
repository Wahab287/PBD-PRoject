```python
#importing Stopword / Tokenization Lib
from nltk.corpus import stopwords 
nltk.download('stopwords')
from nltk.tokenize import word_tokenize 
```

```python
#importing Files Reading And Clustring Lib
import os
import io 
import re
nltk.download('punkt')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from warnings import simplefilter
#imporitng Lib for 
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
#Files or Article Reading
folderpath = r"C:\Users\Hp\Desktop\qw" 
filepaths  = [os.path.join(folderpath, name) for name in os.listdir(folderpath)]
all_files = []

for path in filepaths:
    with open(path, 'r', encoding="ISO-8859-1") as f:
        file = f.readlines()
        all_files.append(file) 
```

```python
#to Check no of Articles
sizeOfmyfiles = len(all_files)
print(sizeOfmyfiles)
```

```python
#convert list to String using join
listToStr = ' '.join(map(str, all_files))
```

```python
#Tokenization + (Excluding Stopwords)
text_tokens = word_tokenize(listToStr)
tokens_without_sw = [word for  word in text_tokens if not word in stopwords.words()]
print(tokens_without_sw)
```

```python
#lemmatization
lemma=WordNetLemmatizer()
lem=[]
for r in tokens_without_sw:
    lem.append(lemma.lemmatize(r))
```

```python
#Removing Special Words
dataupdate=[]
dataupdate=[re.sub('[^a-zA-Z0-9]+','',_) for _ in lem]
dataupdate
```

```python
# K-Mean Clustring 
simplefilter(action='ignore', category=FutureWarning)


vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(dataupdate)

true_k = 5
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print
```

```python
print("Prediction")

Y = vectorizer.transform(["chrome browser to open."])
prediction = model.predict(Y)
print(prediction)

Y = vectorizer.transform(["My cat is hungry."])
prediction = model.predict(Y)
print(prediction)
```
